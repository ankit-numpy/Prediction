```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#reading the dataset
train = pd.read_csv('drive/My Drive/Colab Notebooks/test/zoo.csv')
train
```

```
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
# Transformers
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler

# Modeling Evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score, confusion_matrix, classification_report
```

```
from sklearn.pipeline import Pipeline as pipeline
# Create the X feature matrix and the y target vector
x = train.drop(labels=['class_type','animal_name','species'], axis=1)
y = train['class_type']
# the only step necessary to be done outside of pipeline
# convert the target column to categorical
col = pd.Categorical(y)
y = pd.Series(col.codes)
# split the clean_X into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1, shuffle=True, stratify=y)
```
